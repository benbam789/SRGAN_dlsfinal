```python
import os
from IPython.display import Image, display
import shutil

# Get the current working directory
current_directory = os.getcwd()

# Specify the paths to the train_HR and train_LR folders
train_hr_folder = os.path.join(current_directory, 'custom_dataset/train_HR')
train_lr_folder = os.path.join(current_directory, 'custom_dataset/train_LR')

# Specify the paths to the test_HR and test_LR folders
test_hr_folder = os.path.join(current_directory, 'custom_dataset/test_HR')
test_lr_folder = os.path.join(current_directory, 'custom_dataset/test_LR')

# Create the test_HR folder if it doesn't exist
os.makedirs(test_hr_folder, exist_ok=True)

# Create the test_LR folder if it doesn't exist
os.makedirs(test_lr_folder, exist_ok=True)

# List all files in the train_HR and train_LR folders and sort them alphabetically
train_hr_files = sorted(os.listdir(train_hr_folder))
train_lr_files = sorted(os.listdir(train_lr_folder))

# Get the first 5 pairs of images
first_five_pairs = zip(train_hr_files[:180], train_lr_files[:180])

# Copy the first 5 pairs of images to the test_HR and test_LR folders
for hr_filename, lr_filename in first_five_pairs:
    hr_source_path = os.path.join(train_hr_folder, hr_filename)
    hr_destination_path = os.path.join(test_hr_folder, hr_filename)
    
    lr_source_path = os.path.join(train_lr_folder, lr_filename)
    lr_destination_path = os.path.join(test_lr_folder, lr_filename)
    
    shutil.copy(hr_source_path, hr_destination_path)
    shutil.copy(lr_source_path, lr_destination_path)
```

```python
##GETTinG THE TEST DATA TO CHECK PERFORMANCE METRICS

import os
from PIL import Image
from tqdm import tqdm

# Set your paths
hr_test_hr_folder = 'custom_dataset/hr_test_HR'
gt_data_folder = 'custom_dataset/GT_data'
lr_test_data_folder = 'custom_dataset/LR_test_data'

# Create the GT_data folder if it doesn't exist
os.makedirs(gt_data_folder, exist_ok=True)

# Create the LR_test_data folder if it doesn't exist
os.makedirs(lr_test_data_folder, exist_ok=True)

# Get the list of files in hr_test_HR folder
hr_images = os.listdir(hr_test_hr_folder)[:100]

# Process each image
for img_filename in tqdm(hr_images, desc='Processing images'):
    # Load the HR image
    hr_img_path = os.path.join(hr_test_hr_folder, img_filename)
    hr_img = Image.open(hr_img_path)

    # Save HR image to GT_data folder
    gt_img_path = os.path.join(gt_data_folder, img_filename)
    hr_img.save(gt_img_path)

    # Downsample the HR image by 4x to create LR image
    lr_img = hr_img.resize((hr_img.width // 4, hr_img.height // 4), Image.BICUBIC)

    # Save LR image to LR_test_data folder
    lr_img_path = os.path.join(lr_test_data_folder, img_filename)
    lr_img.save(lr_img_path)

print("Images processed and saved successfully.")
```
